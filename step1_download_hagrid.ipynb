## Step1. HaGRID repo 클론 + 지정 클래스 다운로드

```python
import os
import subprocess
import zipfile
```

### config

```python
# 사용할 클래스 목록
TARGET_CLASSES = [
    "palm",
    "fist",
    "ok",
    "three",
    "like",
    "call",
    "rock",
    "three2",
    "three_gun",  # "gun" → 확인 필요 (위 주석 참고)
]

HAGRID_REPO_DIR = "./hagrid_repo"
RAW_DATA_DIR = "./data/hagrid_raw"
```

### 1. HaGRID repo 클론

```python
def clone_hagrid_repo():
    if os.path.exists(HAGRID_REPO_DIR):
        print(f"[skip] 이미 존재함: {HAGRID_REPO_DIR}")
        return
    print("[1/2] HaGRID 공식 repo 클론 중...")
    subprocess.run(
        ["git", "clone", "https://github.com/hukenovs/hagrid.git", HAGRID_REPO_DIR],
        check=True,
    )
    subprocess.run(
        ["pip", "install", "-r", f"{HAGRID_REPO_DIR}/requirements.txt"],
        check=True,
    )
```

### 2. 필요한 클래스 다운로드

```python
def download_hagrid_classes():
    print(f"[2/2] HaGRID 클래스 다운로드: {TARGET_CLASSES}")
    os.makedirs(RAW_DATA_DIR, exist_ok=True)

    download_script = os.path.join(HAGRID_REPO_DIR, "download.py")

    cmd = [
        "python", download_script,
        "--save_path", RAW_DATA_DIR,
        "--annotations",
        "--dataset",
        "--targets", *TARGET_CLASSES,
    ]
    print(" ".join(cmd))
    subprocess.run(cmd, check=True)
```

```python
if __name__ == "__main__":
    clone_hagrid_repo()
    download_hagrid_classes()
    print("\n다운로드 완료.")
```
