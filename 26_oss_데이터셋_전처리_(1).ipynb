```python
import zipfile
import os

zip_path = "숫자3변형_구글드라이브1.zip"   # 업로드한 zip 파일명으로 수정
extract_folder = "/content/raw_images"

os.makedirs(extract_folder, exist_ok=True)

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_folder)

print("압축 해제 완료:", extract_folder)
```

```python
import os
from PIL import Image, ImageOps
from pillow_heif import register_heif_opener

register_heif_opener()

# ==============================
# 여기만 수정
# ==============================

file_prefix = "number3_modified"

input_folder = "/content/raw_images" #update name
output_folder = "/content/processed_images"

target_size = 640
jpg_quality = 95

# ==============================

valid_extensions = [".jpg", ".jpeg", ".png", ".heic", ".webp", ".bmp"]

os.makedirs(output_folder, exist_ok=True)

# 하위 폴더 안 이미지까지 전부 찾기
image_paths = []

for root, dirs, files in os.walk(input_folder):
    for file in files:
        ext = os.path.splitext(file.lower())[1]
        if ext in valid_extensions:
            image_paths.append(os.path.join(root, file))

image_paths.sort()

print(f"총 {len(image_paths)}개의 이미지 파일을 찾았습니다.")

count = 1

for input_path in image_paths:
    try:
        img = Image.open(input_path)

        # 아이폰/HEIC 사진 방향 보정
        img = ImageOps.exif_transpose(img)

        # RGB로 통일
        img = img.convert("RGB")

        # 비율 유지 resize
        img.thumbnail((target_size, target_size))

        # 640x640 흰 배경 캔버스 생성
        new_img = Image.new("RGB", (target_size, target_size), (255, 255, 255))

        # 가운데 정렬
        x = (target_size - img.width) // 2
        y = (target_size - img.height) // 2
        new_img.paste(img, (x, y))

        # 새 파일명
        new_filename = f"{file_prefix}_{count:02d}.jpg"
        output_path = os.path.join(output_folder, new_filename)

        # JPG로 저장
        new_img.save(output_path, "JPEG", quality=jpg_quality)

        print(f"{os.path.basename(input_path)} → {new_filename}")

        count += 1

    except Exception as e:
        print(f"처리 실패: {input_path}")
        print("에러:", e)

print("전처리 완료!")
print("저장 위치:", output_folder)
```
